## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Copy original merged dataset
df_cleaned = df.copy()

# Remove duplicates
df_cleaned.drop_duplicates(inplace=True)

# Handle known missing values (basic categorical and numerical)
df_cleaned['age_certification'].fillna('Unknown', inplace=True)
df_cleaned['genres'].fillna('Unknown', inplace=True)
df_cleaned['production_countries'].fillna('Unknown', inplace=True)
df_cleaned['runtime'].fillna(df_cleaned['runtime'].median(), inplace=True)
df_cleaned['imdb_score'].fillna(df_cleaned['imdb_score'].mean(), inplace=True)
df_cleaned['imdb_votes'].fillna(0, inplace=True)


# Fill missing TMDB scores/popularity with their mean
df_cleaned['tmdb_score'].fillna(df_cleaned['tmdb_score'].mean(), inplace=True)
df_cleaned['tmdb_popularity'].fillna(df_cleaned['tmdb_popularity'].mean(), inplace=True)

# Fill description with 'Unknown'
df_cleaned['description'].fillna('Unknown', inplace=True)

# Fill 'seasons' (TV Shows only) with 0 and convert to int
df_cleaned['seasons'].fillna(0, inplace=True)
df_cleaned['seasons'] = df_cleaned['seasons'].astype(int)

# Convert data types for analysis
df_cleaned['release_year'] = df_cleaned['release_year'].astype(int)
df_cleaned['runtime'] = df_cleaned['runtime'].astype(int)

# Feature Engineering
# Create a genre list column for genre-based analysis
df_cleaned['genre_list'] = df_cleaned['genres'].apply(
    lambda x: [g.strip() for g in x.split(',')] if x != 'Unknown' else [])

# Create binary flag for content type
df_cleaned['is_movie'] = df_cleaned['type'].apply(lambda x: 1 if x == 'MOVIE' else 0)

df_cleaned.drop(columns=['imdb_id', 'person_id', 'name', 'character', 'role'], inplace=True)

# Final Check
print("Final missing values count (should be 0):", df_cleaned.isnull().sum().sum())
print(f"Final dataset shape: {df_cleaned.shape}")
df_cleaned.head()

# Final check for null values
# null_counts = df_cleaned.isnull().sum()
# null_counts[null_counts > 0]
````

### What all manipulations have you done and insights you found?

## Data Cleaning & Manipulations Summary

To prepare the dataset for analysis, we performed the following data cleaning and transformation steps:

1. **Merged Datasets**: Combined `titles.csv` and `credits.csv` using the common `id` column.
2. **Removed Duplicates**: Used `drop_duplicates()` to ensure no repeated entries.
3. **Handled Missing Values**:
   - Categorical columns (`genres`, `age_certification`, `production_countries`, `description`) filled with `"Unknown"`.
   - Numerical columns (`runtime`, `imdb_score`, `tmdb_score`, `tmdb_popularity`) filled with their **median or mean**.
   - IMDb votes filled with `0` if missing.
   - `seasons` (for shows) filled with `0` and converted to integer.
4. **Dropped Unused Cast Columns**: Since cast/crew info wasn‚Äôt used in analysis, we removed `imdb_id`, `person_id`, `name`, `character`, and `role` which had many missing values.
5. **Converted Data Types**: Ensured `release_year`, `runtime`, and `seasons` are of type `int`.
6. **Feature Engineering**:
   - `genre_list`: Converted comma-separated genre strings into a list for better analysis.
   - `is_movie`: Created a binary flag (1 = Movie, 0 = Show) for content type.

After these steps, **no missing values** remain in the dataset, and it's fully analysis-ready.
